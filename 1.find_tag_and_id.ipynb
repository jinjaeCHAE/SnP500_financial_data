## JavaScript 상의 태그와 ID값을 얻기 위해서

```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# chromedriver.exe 파일이 같은 폴더에 있어야 합니다.
driver = webdriver.Chrome('./chromedriver')

# 크롤링 성공까지 3초 기다려줍니다.
driver.implicitly_wait(3)

# 크롤링 시작합니다.
driver.get('https://finance.yahoo.com/quote/ABBV/financials?p=ABBV')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Yahoo Finance의 경우 구분된 태그와 ID값이 없고 ID값도 지속적으로 바뀝니다. 그래서 크롤링을 할 때, ID값이 다른 이유 등으로 데이터를 잘 가져오지 못합니다.
# 아래 코드를 실행하면 굉장히 많은 데이터가 나옵니다. 여기서 원하는 웹 상의 원하는 데이터를 복사해여 Control + F를 통해 태그와 ID값을 찾습니다.
# 이러한 목적으로 아래 코드를 실행합니다.
print(soup)
```

```python
# 각 태그와 ID가 어떻게 구성되어있는지 파악합니다.
data = soup.find_all(['span'],attrs={'data-reactid':44})
print(data[2].get_text())
# 아래 코드를 실행해보면 하나의 태그에 여러 데이터가 나옵니다. 그래서 크롤러에서 for문으로 모든 데이터를 확인해줘야 합니다.
print(data)
print(len(data))
print(data[0].get_text())
```
